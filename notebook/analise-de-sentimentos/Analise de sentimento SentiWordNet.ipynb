```python
import pandas as pd
import numpy as np
```

```python
# importando os dados
swn = pd.read_csv('./files test/SentiWordNet.csv',sep='	', encoding='ISO-8859-1')
df = pd.read_csv('./files test/tweets_para_teste.csv')
```

```python
#teste
termo = list(swn['Termo'])
valor = list(swn['PosScore'])

for i in range(len(termos)):
    if termo[i] == 'clima':
        print(f'{termo[i]}  {valor[i]}')
```

```python
#busca por valor da palavra no SentiWordNet
def get_valor(palavra, polaridade):
    termo = list(swn['Termo'])
    valor = list(swn[polaridade])
    for i in range(len(termo)):
        if termo[i] == palavra:
            return valor[i]
    return 0


# soma das pontuação dos termos
def get_polaridade(tweet):
    sum_pos = 0
    sum_neg = 0
    for palavra in tweet.split():
        sum_pos += get_valor(palavra,'PosScore')
        sum_neg += get_valor(palavra,'NegScore')
    
    #print(f'pos: {sum_pos}  neg: {sum_neg}')
    if sum_pos > sum_neg:
        print(f'{tweet} -- Positivo')
        return 'positivo'
    elif sum_pos < sum_neg:
        print(f'{tweet} -- Negativo')
        return 'negativo'
    else:
        print(f'{tweet} -- Neutro')
        return 'neutro'
```

```python
#teste
get_polaridade('tudo na vida depende do quanto voce quer comer clima')
```

```python
df['polaridade'] = np.array([get_polaridade(tweet) for tweet in df['TWEET']])
```

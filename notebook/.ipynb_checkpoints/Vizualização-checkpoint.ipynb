```python
def get_LDA(df):
    tp = list(range(0,15))
    dt = df.query(f'topics in {tp}')
    dt = dt.drop_duplicates(subset='topics', keep="last")
    dt = dt.sort_values(by='topics')
    col = ['topics', 'termos']
    col = [c for c in list(dt.columns.values) if c not in col]
    return dt.drop(columns=col)
```

```python
lda = get_LDA(df)
for i in range(len(lda)):
    print(f'Topico {lda.iloc[i][0]}: \n({lda.iloc[i][1]})\n')
```

```python
def plot(df,name):
    percents = [v*100/df['sum'].sum() for i,v in enumerate(df['sum'])]
    polaridade = [v for i,v in enumerate(df['polaridade'])]
    
    pie_chart = pd.DataFrame(percents, index = polaridade, columns=[name])
    fig = pie_chart.plot.pie(subplots=True, autopct='%.2f', figsize=(8, 4))

    #fig = pie[0].get_figure()
    print(type(fig))

    fig

def get_dis_polaridade(df):
    d =  pd.DataFrame(df['polaridade'].value_counts())
    dc = {'polaridade': [], 'sum': []}
    for i in range(len(d)):
        dc['polaridade'].append(d.iloc[i].name)
        dc['sum'].append(d.iloc[i][0])
    return pd.DataFrame(dc)
```

```python
p = plot(get_dis_polaridade(df),'Noticias')
p
```

```python

df = df.dropna(subset=['topics'])
comlunms = list(df.columns.values)
c = ['topics', 'termos', 'polaridade']
comlunms = [x for x in comlunms if x not in c]
df = df.drop(columns=comlunms)
topics = pd.DataFrame(df.groupby(['topics']))
topics.columns = ['topico', 'seila']
topics = [topics['topico'].iloc[x] for x in range(len(topics.topico))]
for topico in topics:
    df_top = df[df['topics'] == topico]
    #print('=='*15)
    #print(f"Topico {int(topico)} \n({df_top['termos'].iloc[2]})")
    termos = df_top['termos'].iloc[2]

    df_top = pd.DataFrame(df_top['polaridade'].value_counts())

    dc_topicos = {'polaridade':[], 'quantidade':[]}
    for i in range(len(df_top)):
        dc_topicos['polaridade'].append(df_top.iloc[i].name)
        dc_topicos['quantidade'].append(df_top.iloc[i][0])

       
    df_top = pd.DataFrame(dc_topicos)

    # conjunto de dados
    porcentagem = [list(df_top['quantidade'])]
    polaridade = list(df_top['polaridade'])

    percents = [v*100/df_top['quantidade'].sum() for i,v in enumerate(df_top['quantidade'])]
    polaridade = [v for i,v in enumerate(df_top['polaridade'])]


    dc_color = {'Positivo':'#32CD32', 'Negativo':'#B22222', 'Neutro':'#4682B4'}
    colors = [dc_color[i] for i in polaridade]

    
    label = f"Topico {int(topico)} \n({termos})"
    pie_chart = pd.DataFrame(percents, index = polaridade)
    fig = pie_chart.plot.pie(subplots=True, autopct='%.2f', figsize=(8, 4), colors=colors, title=label)
    fig

    
```

## Word Cloud

```python
stopwords = set(STOPWORDS)

words = ' '.join(df['text'])
word_clean = " ".join([word for word in words.split()
                      if 'https' not in word
                      and 'href' not in word
                      and 'rel' not in word
                      and 'nofollow' not in word
                      and 'mixvale' not in word])
```

```python
warnings.simplefilter('ignore')
#twitter_mask = imread('brasil_mask.png')

wc = WordCloud(min_font_size=10,
              max_font_size=300,
              background_color='white',
              mode="RGB",
              stopwords=stopwords,
              width=2000,
              height=1000,
              #mask=twitter_mask,
              normalize_plurals = True).generate(word_clean)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
